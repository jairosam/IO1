## 1.
### Resolver
* $2x+y<5$

### Solución
$$2x+y<5$$
$$y<5-2x$$
$$y<-2x+5$$
Donde:
$$m = -2$$
$$b = 5$$



```python
def y(x):
    return -2*x+5
```

```python
puntos_x = np.arange(-5,5,1)

puntos_y = [y(x) for x in puntos_x]
```

```python
plt.grid(True)
plt.axhline(0,color = 'k',lw =2)
plt.axvline(0,color = 'k',lw =2)
plt.plot(puntos_x, puntos_y, label="y=-2x+5")
plt.fill_between(puntos_x, puntos_y,np.min(puntos_y), alpha=0.3, color='red')
plt.legend()
```

## 3.
### Resolver
* $2(2x-y)<2(x+y)-4$

### Solución
$$2(2x-y)<2(x+y)-4$$
$$4x-2y<2x+2y-4$$
$$2x+4<4y$$
$$\dfrac{1}{2}x+1<y$$
Donde:
$$m = \dfrac{1}{2}$$
$$b = 1$$

```python
def y(x):
    return (1/2)*x + 1
```

```python
puntos_x = np.arange(-5,5,1)

puntos_y = [y(x) for x in puntos_x]
```

## 4.
### Resolver 
* $2x+y>3$
* $2y-1>0$
* $y\leq x$

### Solución
$$2x+y>3$$
$$y>3-2x$$


$$2y-1>0$$
$$2y>1$$
$$y>\dfrac{1}{2}$$


$$y\leq x$$

```python
def y1(x):
    return 3 - 2*x
def y2(x):
    return 1/2
def y3(x):
    return x
```

```python
puntos_x1 = np.arange(-5,7,1)
puntos_x2 = np.arange(-5,7,1)
puntos_x3 = np.arange(-5,7,1)

puntos_y1 = [y1(x) for x in puntos_x1]
puntos_y2 = [y2(x) for x in puntos_x2]
puntos_y3 = [y3(x) for x in puntos_x3]
puntos_y4 = np.maximum(puntos_y3, puntos_y2)
```

```python
plt.grid(True)
plt.axhline(0,color = 'k',lw =2)
plt.axvline(0,color = 'k',lw =2)
plt.plot(puntos_x1, puntos_y1, label="y=3-2x")
plt.plot(puntos_x2, puntos_y2, label="y=1/2")
plt.plot(puntos_x3, puntos_y3, label="y=x")

plt.fill_between(puntos_x1,puntos_y4,puntos_y2,where=puntos_y4>=puntos_y1,alpha=0.3, color='red')
plt.legend()
```

## 6. Resolver 
Maximizar
* $P =4x+6y$ 

Sujeto a:
* $2x+y \leq 180$
* $x+2y \leq 160$
* $x+y \leq 100$
* $x \geq 0$, $y \geq 0$

Despejamos estas ecuaciones y graficamos

* $y=180-2x$
* $y=\dfrac{160-x}{2}$
* $y=100-x$

```python
def res1(x):
    return 180-2*x

def res2(x):
    return (160-x)/2

def res3(x):
    return 100-x 
```

```python
import numpy as np

X1 = np.arange(0,99,9)
X2 = np.arange(0,176,16)
X3 = np.arange(0,110,10)
y1 = [res1(x) for x in X1]
y2 = [res2(x) for x in X2]
y3 = [res3(x) for x in X3]
```

```python
plt.plot(X1,y1,label="Restricción 1")
plt.plot(X2,y2,label="Restricción 2")
plt.plot(X3,y3,label="Restricción 3")
plt.legend()
plt.grid(True)
puntos_corte = np.array([[0,80],[40,60],[80,20],[90,0]])
for punto in puntos_corte:
    plt.scatter(punto[0],punto[1],c="yellow") 
plt.axhline(0,color = 'k',lw =2)
plt.axvline(0,color = 'k',lw =2)
plt.fill_between(puntos_corte[:,0],puntos_corte[:,1], alpha=0.3, color='red')
```

```python
def objetivo(x,y):
    return 4*x + 6*y

evaluar = [objetivo(punto[0],punto[1]) for punto in puntos_corte]
print("""
    Evaluaciones de los puntos de corte: {}
    Valores de "x" y "y" más optimos: {}
    Resultado más optimo: {}
""".format(evaluar, puntos_corte[evaluar.index(max(evaluar))],max(evaluar)))
```

### Problema

Una compañia fabrica 2 productos $X$ e $Y$. c/u de estos requiere cierto tiempo en la linea de ensamblado y en la linea de acabado. Cada articulo del tipo $X$ necesita 5h de Ensamblado y 2h de Acabado cada uno, del tipo $Y$ necesita 3h de Ensamblado y 4h de acabado. En cualquier semana, la empresa dispone de:

* 105h en la linea de ensamblado
* 70h en la linea de acabado

La empresa puede vender todos los articulos y obtener una utilidad de:

* 200 por cada uno de los articulos de $X$
* 100 por cada uno de los articulos de $Y$

Determinar el N° de articulos de cada tipo que se debe fabricar para maximizar la utilidad

### Solución

Se han planteado las siguientes ecuaciones: 

$$Z = 200x + 100y$$

Con las siguientes restricciones: 

$$5x+3y \leq 105$$
$$2x + 4y \leq 70$$
$$x \geq 0, \ y \geq 0$$

Al despejar $y$ de las restricciones obtenemos:

$$y \leq \dfrac{105-5x}{3}$$

$$y \leq \dfrac{70-2x}{4}$$

```python
def res1(x):
    return (105-5*x)/3

def res2(x):
    return (70-2*x)/4
```

```python
datos_1 = np.arange(0,22,1)
datos_2 = np.arange(0,36,1)

y_res1 = [res1(x) for x in datos_1]
y_res2 = [res2(x) for x in datos_2]
```

```python
plt.plot(datos_2, y_res2, label="Restricción 1")
plt.plot(datos_1, y_res1, label="Restricción 2")
plt.legend()
puntos_corte = np.array([[0,17.5],[15,10],[21,0]])
for punto in puntos_corte:
    plt.scatter(punto[0],punto[1],c="yellow")
plt.grid(True)
plt.fill_between(datos_2[:16], y_res2[:16], alpha=0.3, color='red')
plt.fill_between(datos_1[15:], y_res1[15:], alpha=0.3, color='red')
plt.axhline(0,color = 'k',lw =2)
plt.axvline(0,color = 'k',lw =2)
```

```python
resultados = []
def objetivo(x,y):
    return 200*x + 100*y

for punto in puntos_corte:
    resultados.append(objetivo(punto[0], punto[1]))

print("""
    Evaluaciones de los puntos de corte: {}
    Valores de "x" y "y" más optimos: {}
    Resultado más optimo: {}
""".format(resultados, puntos_corte[resultados.index(max(resultados))],max(resultados)))
```

Por tanto el resultado más optimo será: 

* El número de articulos $X$ que se debe fabricar es de 21
* El número de articulos $Y$ que se debe fabricar es de 0
* La utilidad maximizada es de 4200

### Ejercicio método gráfico (máx.)
## problema:
* soldados y trenes. Un solo soldado se vende en 27 y requiere de 10 de materia prima. Cada soldado que se fabrica requiere de costos adicionales en 14. Un tren se vende en 21 y se utiliza 9 de su valor en materia prima. Todos los trenes tienen costos adicionales de 10  por tren.Se requiere dos tipos de mano de obra especializada: carpintería y acabados. Un soldado necesita dos horas de trabajo da acabado y una hora de carpintería. Un tren requiere una hora de acabado y una de carpintería. Se consigue todo el material, pero solo tiene 100 horas de trabajo de acabado y 80 de carpintería. Se vende cuando menos de 40 soldados por semana y a lo mas 30 trenes. Se desea maximizar las utilidades semanales. 

## solución:

 
Siendo	$x =$soldado	$y =$tren 

se plantea el siguiente modelo a MAXIMIZAR

* $ z = 3x + 2y $ (funcion objetivo)	

sujeto a las  restricciones

* $ x + y \leq 80 $

* $ x \geq 40$

* $ 2x + y \leq 100 $

* $ y \leq 30 $

* $ x,y \geq 0 $



```python
def function1(x):
    return (80-x)

def function2(x):
    return (100-2*x)

def function3(y):
    return 40

def function4(x):
    return 30

def focusFunction(x,y):
    z = (3*x)+(2*y)
    return z
```

```python
dataX1 = np.arange(0,80,1)
dataX2 = np.arange(0,51,1)
dataY3 = np.arange(0,100,1)
dataX4 = np.arange(0,80,1)

dataY1 = [function1(x) for x in dataX1]
dataY2 = [function2(x) for x in dataX2]
dataX3 = [function3(y) for y in dataY3]
dataY4 = [function4(x) for x in dataX4]
```

```python
plt.plot(dataX1, dataY1, label="Restricción 1")
plt.plot(dataX2, dataY2, label="Restricción 2")
plt.plot(dataX3, dataY3, label="Restricción 3")
plt.plot(dataX4, dataY4, label="Restricción 4")
plt.axhline(0,color = 'k',lw =2)
plt.axvline(0,color = 'k',lw =2)
plt.legend()

puntos_corte = np.array([[40,0],[40,20],[50,0]])
for punto in puntos_corte:
    plt.scatter(punto[0],punto[1],c="yellow")

plt.fill_between(puntos_corte[:,0], puntos_corte[:,1],alpha=0.3, color="red")    

dataZ = [focusFunction(40,20),focusFunction(40,0),focusFunction(50,0)]
print(max(dataZ))
```

* se dice que si la manufactura vende 40 juguetes soldados y 20 juguetes trenes la ganancia máxima que obtendrá es de 160.

### Ejercicio método gráfico (mín.)
## problema:

Un expendio de carnes acostumbra preparar carne para hamburguesa con una combinación de carne molida de res y carne molida de cerdo. La carne de res contiene 80 % de carne y 20 % de grasa y le cuesta a la tienda 80 centavos por libra. La carne de cerdo contiene 68 % de carne y 32 % de grasa y cuesta 60 centavos por libra. 
* ¿Qué cantidad de cada tipo de carne debe emplear la tienda por cada libra de carne para hamburguesa si desea minimizar el costo y mantener el contenido de grasa no mayor de 25 %?

## solución:

 
Siendo:
* $x =$carne de res	
* $y =$carne de cerdo

se plantea el siguiente modelo a MINIMIZAR

* $ z = 80x + 60y $ (funcion objetivo)	

sujeto a las  restricciones

* $ 20x + 32y \leq 25 $

* $ x + y = 1 $

* $ x,y \geq 0 $



```python
def function1(x):
    return (25-20*x)/32

def function2(x):
    return (1-x)

def focusFunction(x,y):
    return(80*x+60*y)
```

```python
dataX1 = np.arange(0,1.5,1)
dataX2 = np.arange(0,2,1)

dataY1 = [function1(x) for x in dataX1]
dataY2 = [function2(x) for x in dataX2]
```

```python
plt.plot(dataX1, dataY1, label="Restricción 1")
plt.plot(dataX2, dataY2, label="Restricción 2")
plt.axhline(0,color = 'k',lw =2)
plt.axvline(0,color = 'k',lw =2)
plt.legend()

plt.fill_between([0,7/12], np.max(dataX2), [1,5/12], alpha=0.3,color='red')
plt.fill_between([7/12,1], np.max(dataX1), [5/12,dataY1[len(dataY1)-1]], alpha=0.3,color='red')

plt.scatter(7/12,5/12,c='red')
print(focusFunction(7/12,5/12))
```

* $x =$ 7/12 lbs de carne de res
* $y =$ 5/12 lbs de carne de cerdo
* $z =$ 215/3 centavos
